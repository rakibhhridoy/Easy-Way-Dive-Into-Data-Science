## Regression Plots

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips, marker='x', color= 'red')

# the line is showing us the regression line that reflects the two variables in common path..
```

```python
sns.regplot(x = 'tip', y = 'total_bill', data = tips, scatter_kws={
    'marker' : 'x',
    'color' : 'purple'
}, 
           line_kws={
               'linewidth' : 2,
               'color': 'blue'
           })
```

```python
# estamating standard error with ci

sns.regplot(x = 'total_bill', y = 'tip', data= tips, marker = 'x', color= 'purple', line_kws= {'color' : 'blue', 'linewidth' : 3.2}, ci = .45)
```
